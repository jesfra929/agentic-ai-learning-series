## **üß† Configuring LLMs in CrewAI (OpenAI + other providers)**

### **1. What happens by default?**

If you **only set `OPENAI_API_KEY`** in your environment and don‚Äôt specify any LLM:

```python
from crewai import Agent

agent = Agent(
    role='Your Role',
    goal='Your Goal',
    backstory='Your Backstory'
    # No llm parameter
)
```

* CrewAI will automatically use **OpenAI** üß©
* The **default model** is: `gpt-4o-mini`
* You **don‚Äôt need** to create an `LLM` object or pass `llm=...` for this default behavior.

**Use this when:**

* You‚Äôre okay with the default OpenAI model (`gpt-4o-mini`)
* You just want things to ‚Äúwork‚Äù with minimal config

<br>

### **2. Changing the OpenAI model**

If you want a **different OpenAI model** (e.g. `gpt-4`, `gpt-4o`, `gpt-4-turbo`, `o1`, `o1-mini`), you have two options.

#### üè∑Ô∏è **Option A ‚Äì Pass model as a simple string (quick & simple)**

```python
from crewai import Agent

openai_agent = Agent(
    role='OpenAI Expert',
    goal='Provide insights using GPT-4',
    backstory="An AI assistant powered by OpenAI's latest model.",
    llm='gpt-4'  # simple string, OpenAI assumed
)
```

#### üè∑Ô∏è **Option B ‚Äì Use the `LLM` class (recommended for control)**

```python
from crewai import Agent, LLM

llm = LLM(
    model="openai/gpt-4",   # note the provider prefix
    temperature=0.7
)

agent = Agent(
    role='Your Role',
    goal='Your Goal',
    backstory='Your Backstory',
    llm=llm
)
```

**Best practice:** when using the `LLM` class, **always include the provider prefix**, like `openai/`.

<br>

### **3. Common OpenAI models**

* `openai/gpt-4o` ‚Äì Optimized for speed + intelligence
* `openai/gpt-4-turbo` ‚Äì Great for long-form content
* `openai/gpt-4` ‚Äì High-accuracy tasks
* `openai/o1`, `openai/o1-mini` ‚Äì Advanced reasoning models

‚ö†Ô∏è **Important:**

> Always include the `openai/` prefix when using the `LLM` class (e.g. `model="openai/gpt-4"`).

<br>

### **4. Using non-OpenAI models (e.g. Gemini)**


#### üìç **A. With `.env`‚Äìstyle API key**

## **üìù Pre-Configured LinkedIn Post Creator (Gemini ‚Äì No User Input)**

This setup creates a fully pre-configured AI agent using Google‚Äôs Gemini model to generate motivational LinkedIn posts for job seekers. All instructions, style, tone, and output requirements are embedded directly within the `Task` class, so no additional user input is required during execution.

````python
from crewai import Agent, Task, Crew, LLM

llm = LLM(
    model ="gemini-2.5-flash-lite",
    api_key=GOOGLE_API_KEY
)

agent = Agent(
    role = "Empathetic Career Coach",
    goal = "Craft a powerful and encouraging message for job seekers, igniting their motivation and confidence",
    backstory = "An experienced mentor who inspires others through their insightful writings and articles.",
    llm = llm
)
````

````python
task = Task(
    description = """
    Craft a concise, highly engaging, and profoundly motivating LinkedIn post for job seekers.
    Infuse the message with light humor and a fresh, 'eye-opening' perspective.
    Make it impactful by integrating recent incidents, real-time examples, or a compelling inspiring story (real or fictional).
    Start with a powerful, attention-grabbing hook to immediately captivate the audience and ignite their confidence.
    The post should also incorporate clean, neat headings, subheadings, bullet points, and appropriate emojis to enhance readability and make it look premium and professional.
    """,
    agent = agent,
    expected_output = "A 3-4 paragraph LinkedIn post, approximately 150-250 words, structured with an engaging hook, a humorous or eye-opening anecdote/example, motivational message, and a clear call to action, suitable for a professional network, featuring clean headings, subheadings, bullet points, and relevant emojis."
)
````

````python
crew = Crew(
    agents = [agent],
    tasks = [task]
)
````

````python
print(crew.kickoff())
````

## **üìù Dynamic LinkedIn Post Creator (With Custom Topic Input)r**

In this section, we build an AI agent that generates motivational LinkedIn posts for job seekers using dynamic input passed through variables in the CrewAI `Task` configuration. A custom topic is provided at runtime using `crew.kickoff()`, enabling flexible and personalized content creation.


### üìò **Understanding Variable Usage in CrewAI**

Variables allow you to dynamically customize the behavior, tone, and output of agents and tasks **without modifying the core code**. This makes your workflow more **flexible, reusable, and scalable**.

<br>

#### üéØ **Why Use Variables?**

* Personalize outputs based on different contexts
* Reuse the same agent/task for multiple use cases
* Avoid rewriting long prompts repeatedly
* Improve maintainability and reduce errors

<br>

#### üß© **Where Can Variables Be Used?**

You may insert variables inside:

üîπ **Agent Attributes**

* `role`
* `goal`
* `backstory`

üîπ **Task Fields**

* `description`
* `expected_output`

Example placeholder format:

```
{{ variable_name }}
```
<br>

#### üöÄ **How to Supply Variable Values**

Use `crew.kickoff()` to pass values dynamically:

```python
crew.kickoff(
    inputs={
        "variable_name": "Your Value"
    }
)
```
<br>

#### üìè **Rules & Best Practices**

* Always wrap variables using `{{ ... }}`
* Variable names must match exactly between the code and `inputs`
* Use meaningful, descriptive names (ex: `topic`, `audience_type`, `years_experience`)
* Only convert text to variables **if you expect it to change**
* Keep instructions clear and neutral ‚Äî avoid ambiguity

<br>

#### üìù **Quick Summary**

‚úî Variables make AI **dynamic**
<br>
‚úî Use `{{ variable }}` placeholders
<br>
‚úî Supply values via `inputs={...}`
<br>
‚úî Works in **Agent** and **Task**
<br>
‚úî Helps build **smart, reusable notebooks**

## **Human in the loop + Verbose**


### üìå **About This Agent** üîΩ

This demo shows how an AI agent **appears to remember** the user's name and then uses it to **generate a personalized, fun notification message**.

The flow works in **two tasks without enabling CrewAI‚Äôs official memory module**:

1Ô∏è‚É£ **Task 1:** The agent receives the user‚Äôs name.
2Ô∏è‚É£ **Task 2:** The agent produces a **funny, app-style notification** using the same name ‚Äî based on **LLM contextual memory**, not configured memory settings.

‚ö†Ô∏è **Note:**
This is **not real persistent memory** ‚Äî the agent is remembering only through the **LLM‚Äôs natural short-term context**.
In the **next cell**, we will enable **actual CrewAI memory (`memory=True`)** to show the difference.



````python
from re import VERBOSE
from crewai import Agent, Task, Crew

memory_demo_agent = Agent(
    name="PlayfulMemoryAgent",
    role="A friendly notification-style AI that remembers the user's name",
    goal=(
        "Remember the user's name when provided, and later use it to create "
        "fun notification-style messages without asking for it again."
    ),
    backstory=(
        "You are a fun, friendly, and playful AI designed to interact like modern "
        "mobile app notifications. You can remember the user's name once they provide it "
        "and must reuse it in future tasks without requesting it again."
    ),
    llm = llm,
)


# Task 1:Store user's name in memory
remember_username_task = Task(
    description="The user name is {{user_name}}, remember this for future messages.",
    expected_output="A confirmation that the name is remembered.",
    agent=memory_demo_agent
)

# Task 2:Generate personalized Notification
memory_demo_task = Task(
    description=(
        "You have already remembered the user's name from the previous task. "
        "Now, using that remembered name, generate EXACTLY ONE notification-style message. "
        "The message should be unique and creative, suitable for a mobile app notification, "
        "and must include the remembered user's name. "
        "Start the message with a category (e.g., 'Food App Notification:', 'Gaming App Notification:', 'Rider App Notification:'). "
        "For example: 'Food App Notification: Promise me <name>, don‚Äôt keep your tummy hungry! ü§ùüç±' or 'Gaming App Notification: <name>, new mission unlocked: Stay happy and hydrated! üéÆüíß'\n\n"
        "Replace <name> ONLY with the remembered user name. "
        "Do not greet, do not explain, do not ask for the name again, "
        "and do not output anything else outside the required single line message."
    ),
    expected_output="Exactly one unique, notification-style formatted line using the remembered user name, starting with a category like 'Food App Notification:'.",
    agent=memory_demo_agent
)
````

````python
snackbar_crew = Crew(
    agents=[memory_demo_agent],
    tasks=[remember_username_task, memory_demo_task],
)


user_name = input("Hi! What is your name? \n")

result = snackbar_crew.kickoff(inputs={"user_name":user_name})
print(result)
````

## **Memory**


### üß† **Understanding Memory in CrewAI (Simple Guide)**

When you enable memory in CrewAI using:
```
memory=True
```


You **do NOT need to explicitly mention** short-term, long-term, or entity memory.

CrewAI automatically enables:

* **Short-term memory**
* **Long-term memory**
* **Entity memory**

All with **default settings**, working transparently out of the box.

You only configure them manually if you need **custom behavior** such as:

* Different embedding providers
* Custom storage paths
* Using Mem0 / Qdrant / Chroma
* Per-memory embedding models

<br>

### üîß **1. Global (Shared) Embedder for All Memory Types**

If you specify an embedder at the **Crew level**, that embedder is shared by
**short-term, long-term, and entity memory automatically**.

This is the simplest and most common setup.

```python
from crewai import Crew

crew = Crew(
    agents=[agent],
    tasks=[task],
    memory=True,
    embedder={
        "provider": "anthropic",       # Match your LLM provider
        "config": {
            "api_key": "your-anthropic-key",
            "model": "text-embedding-3-small"
        }
    }
)
```

‚úî Easiest setup

‚úî No manual memory wiring

‚úî One embedder used across all memory types

<br>

### üéØ **2. Custom Embedder for Each Memory Type (Advanced)**

Use this only if you want
**different providers/models** for short-term, long-term, and entity memory.

```python
from crewai import Crew
from crewai.memory.short_term.short_term_memory import ShortTermMemory
from crewai.memory.long_term.long_term_memory import LongTermMemory
from crewai.memory.entity.entity_memory import EntityMemory

# Custom embedder for short-term memory
short_term_memory = ShortTermMemory(
    embedder_config={
        "provider": "openai",
        "config": {"model": "text-embedding-3-small"}
    }
)

# Custom embedder for long-term memory
long_term_memory = LongTermMemory(
    embedder_config={
        "provider": "cohere",
        "config": {"model": "embed-english-v3.0"}
    }
)

# Custom embedder for entity memory
entity_memory = EntityMemory(
    embedder_config={
        "provider": "google",
        "config": {"model": "models/embedding-001"}
    }
)

crew = Crew(
    agents=[...],
    tasks=[...],
    memory=True,
    short_term_memory=short_term_memory,
    long_term_memory=long_term_memory,
    entity_memory=entity_memory
)
```

‚úî Maximum customization

‚úî Each memory type can use a different embedder

‚úî Ideal for large or specialized agents

````python
from sentence_transformers import SentenceTransformer

local_model = SentenceTransformer('all-MiniLM-L6-v2')

embedder_config = {
    "provider": "local",
    "config": {
        "embedder_instance": local_model
    }
}
````

````python
import os
from crewai import Crew, Agent, Task, Process, LLM
from google.colab import userdata

# Set your Google API key
GOOGLE_API_KEY = userdata.get('GEMINI_API_KEY')
os.environ["GOOGLE_API_KEY"] = GOOGLE_API_KEY
os.environ["GEMINI_API_KEY"] = GOOGLE_API_KEY

# Create Agents
researcher = Agent(
    role="Research Analyst",
    goal="Gather comprehensive information on a given topic.",
    backstory="Expert in online research and information synthesis.",
    verbose=True,
    llm=LLM(model="gemini/gemini-2.0-flash",api_key=GOOGLE_API_KEY)
)

writer = Agent(
    role="Content Writer",
    goal="Create engaging and informative content based on research.",
    backstory="Skilled in writing articles and reports.",
    verbose=True,
    llm=LLM(model="gemini/gemini-2.0-flash",api_key=GOOGLE_API_KEY)
)

# Create Tasks
research_task = Task(
    description="Research the latest advancements in renewable energy technologies.",
    agent=researcher,
    expected_output="Researched information about the topic"
)

write_task = Task(
    description="Write a concise report summarizing the key advancements in renewable energy, based on the research provided.",
    agent=writer,
    expected_output="Concise information about the topic"
)


embedder_config = {
    "provider": "sentence-transformer",
    "config": {
        "model": "all-MiniLM-L6-v2"
    }
}

my_crew = Crew(
    agents=[researcher, writer],
    tasks=[research_task, write_task],
    process=Process.sequential,
    memory=True,
    verbose=True,
    embedder=embedder_config
)
````

````python
result = my_crew.kickoff()
print(result)
````
